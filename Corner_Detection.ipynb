```python
import cv2 as cv
from matplotlib import pyplot as plt
import numpy as np
```

```python
flat_chess = cv.imread('photos/chess.png')
flat_chess = cv.cvtColor(flat_chess , cv.COLOR_BGR2RGB)
plt.imshow(flat_chess)
```

```python
gray_flat = cv.cvtColor(flat_chess , cv.COLOR_RGB2GRAY)
plt.imshow(gray_flat , cmap='gray')
```

```python
real_chess = cv.imread('photos/real_chess.png')
real_chess = cv.cvtColor(real_chess , cv.COLOR_BGR2RGB)
plt.imshow(flat_chess)
```

```python
gray_real = cv.cvtColor(real_chess , cv.COLOR_RGB2GRAY)
plt.imshow(gray_real , cmap='gray')
```

```python
gray = np.float32(gray_flat)
```

```python
dst = cv.cornerHarris(src=gray , blockSize=2 , ksize=3 , k=0.04)
dst = cv.dilate(dst,None)
flat_chess[dst > 0.01* dst.max()] = [255,0,0] 
plt.imshow(flat_chess)
```

```python
gray = np.float32(gray_real)
dst = cv.cornerHarris(src=gray , blockSize=2 , ksize=5 , k=0.04)
dst = cv.dilate(dst,None)
real_chess[dst > 0.01* dst.max()] = [255,0,0] 
plt.imshow(real_chess)
```

```python
flat_chess = cv.imread('photos/chess.png')
flat_chess = cv.cvtColor(flat_chess , cv.COLOR_BGR2RGB)

real_chess = cv.imread('photos/real_chess.png')
real_chess = cv.cvtColor(real_chess , cv.COLOR_BGR2RGB)

gray_flat = cv.cvtColor(flat_chess , cv.COLOR_RGB2GRAY)
gray_real = cv.cvtColor(real_chess , cv.COLOR_RGB2GRAY)
```

```python
#STEP2 
corners = cv.goodFeaturesToTrack(gray_flat , 150 , 0.04 ,1)
corners = np.intp(corners)

for i in corners:
    x,y = i.ravel()
    cv.circle(flat_chess , (x,y), 3 , (255,0,0) , -1)
plt.imshow(flat_chess)
```

```python
corners = cv.goodFeaturesToTrack(gray_real , 200, 0.04 ,1)
corners = np.intp(corners)

for i in corners:
    x,y = i.ravel()
    cv.circle(real_chess , (x,y), 3 , (255,0,0) , -1)
plt.imshow(real_chess)
```
